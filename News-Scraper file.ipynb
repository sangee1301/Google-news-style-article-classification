```python
#import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import csv

import feedparser as fp
import numpy as np
import json
import newspaper
import time
from newspaper import Article
from time import mktime
from datetime import datetime

from tqdm import tqdm
```

Loading sites from a json file

```python
#Limit of no. of atricles that can be download
LIMIT = 50
articles_array = []

data = {}
data['newspapers'] = {}

# Loads the JSON files with the news site links
with open('sites.json') as data_file:
    file = json.load(data_file)
```

Downloading articles using python newspaper library

```python
count = 1

# Iterate through each site name
for site, value in tqdm(file.items()): 
# Extracting articles using pyhton newspaper library
    print("Downloading articles from ", site)
    paper = newspaper.build(value['link'], memoize_articles=False)
    newsPaper = {
        "link": value['link'],
        "articles": []
    }
    noneTypeCount = 0
    for content in paper.articles:
        if count > LIMIT:
            break
        try:
            content.download()
            time.sleep(1)
            content.parse()
        except Exception as e:
            print(e)
            print("continuing...")
            continue
        article = {}
        article['category'] = site
        article['title'] = content.title
        article['authors'] = content.authors
        article['text'] = content.text
        article['link'] = content.url
        article['published'] = content.publish_date
        newsPaper['articles'].append(article)
        articles_array.append(article)
        print(count, content.url)
        count = count + 1
            
    count = 1
    data['newspapers'][site] = newsPaper
```

Saving the scrapped articles to a CSV file

```python
#Saving the scrapped details in a CSV file
try:
    f = csv.writer(open('news_output.csv', 'w', encoding='utf-8'))
    f.writerow(['siteName','Title','Authors','Text','Link','Published_Date'])
    #print(article)
    for article_name in articles_array:
        siteName = article_name['category']
        title = article_name['title']
        authors=article_name['authors']
        text=article_name['text']
        #image=article_name['top_image']
        #video=article_name['movies']
        link=article_name['link']
        publish_date=article_name['published']
        # writing article details as rows
        f.writerow([siteName, title, authors, text,link, publish_date])
except Exception as e: print(e)
```
